```python
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
import numpy as np
```

Barplot using <b>State-UT-wise Funds Allocated, Released and Utilized under Pradhan Mantri Awaas Yojana-Gramin from 2016-17 to 2018-19</b> dataset, for year 2016-17 and 2017-18.

```python
# read dataset
data = pd.read_excel('State-UT-wise Funds Allocated, Released and Utilized under Pradhan Mantri Awaas Yojana-Gramin from 2016-17 to 2018-19.xls')
```

```python
data.head()
```

```python
plt.figure(figsize=(16,10))
plt.subplot(121)
sn.barplot(x = '2016-17 - Utilization *', y = 'State/UT', data = data,
            label = '2016-2017 Utilization', color = 'b', edgecolor = 'w')
plt.xlabel("in Lakhs",fontweight='bold')
plt.ylabel("State/UT",fontweight='bold')
plt.title("Bar Plot for State wise fund Utilization for 2016-17",fontweight='bold')
plt.legend(ncol = 1, loc = 'upper right')
sn.despine(left = True, bottom = True)
plt.subplot(122)
sn.barplot(x = '2017-18 - Utilization *', y = 'State/UT', data = data,
            label = '2017-2018 Utilization', color = 'b', edgecolor = 'w')
plt.xlabel("in Lakhs",fontweight='bold')
plt.ylabel("State/UT",fontweight='bold')
plt.title("Bar Plot for State wise fund Utilization for 2017-18",fontweight='bold')
plt.legend(ncol = 1, loc = 'upper right')
sn.despine(left = True, bottom = True)
plt.tight_layout()
plt.savefig('BarPlot.png', format='png', dpi = 300, bbox_inches = 'tight')
```

For scatter plot we used <b>total-number-and-amount-money-orders-india-during-2000-01-2010-11</b> dataset. Scatter plot is being plotted between Year and Number of Indian moneyorders issued 

```python
#data read
data = pd.read_csv('total-number-and-amount-money-orders-india-during-2000-01-2010-11.csv')
```

```python
data.head()
```

```python
# scatter plot for "year" vs " number of money order"
plt.figure(figsize=(10,5))
sn.scatterplot(x = 'Year',y='Indian Money Orders Issued and paid in India (in lakh)',data= data)
plt.xlabel("Year",fontweight='bold')
plt.title("Scatter Plot for Number of Indian money orders issued",fontweight='bold')
plt.ylabel("Indian Money Orders Issued and paid in India (in lakh)",fontweight='bold')
plt.grid()
plt.savefig('ScatterPlot.png', format='png', dpi = 300, bbox_inches = 'tight')
```

For box plot we used <b> water-quality-tributary-streams-panchganga-and-bhima-2014</b> dataset. The boxplot is being plotted between STATES and pH value.

```python
data = pd.read_excel('water-quality-tributary-streams-panchganga-and-bhima-2014.xls')
```

```python
data.head()
```

```python
plt.figure(figsize=(10,5))
sn.boxplot(x='STATE',y='pH  : Min  : 6.5-8.5',data=data)
plt.xlabel("State",fontweight='bold')
plt.ylabel("pH Min Value",fontweight='bold')
plt.title("Box Plot between States and pH Min Value",fontweight='bold')
plt.savefig('BoxPlot.png', format='png', dpi = 300, bbox_inches = 'tight')
```
